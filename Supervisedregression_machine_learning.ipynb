##Regression machine learning
It is a supervised learning technique because the dataset used contains some features and outcomes(label) Regression focuses on

```python
# Sales prediction based on advertising methods
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the sales dataset
sales_data=pd.read_csv('https://msi.martial.co.ke/data/Advertising.csv')
sales_data.head(5)
```

```python
sales_data['Outcome']=sales_data['Sales'] * 2
sales_data.head(5)
```

```python
# Data analysis
sales_data.info()
```

```python
# Checking empties
sales_data.isnull().sum()
```

```python
# Statistical analysis
sales_data.describe()
```

```python
# correlation analysis
# feature elimination- removing any feature that contributes to a negative
sales_data.corr()
```

```python
# Machine learning
# 1. Divide data sets into features(x) and labels(y)
# convert the sales DataFrame into Numpy arrays so that we can extract its values
array=sales_data.values
array
# x-features(Tv,Radio,Newspaper)
x=array[:, 0:3]
print(x.shape)
y=array[:,3]
print(y.shape)
```

```python
# 2. To divide x and y into training(70%) and testing(30%)
# x-> x_train(70%) and x_test(30%)
# y-> y_train(70%) and y_test(30%)
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(x, y, test_size=0.30, random_state=42)
print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

```python
# 3.Importing scientific machine learning algorithms
# Linear regressions
from sklearn.linear_model import LinearRegression
# b) Decision Trees
from sklearn.tree import DecisionTreeRegressor
# c) K-Nearest neighbour algorithms
from sklearn.neighbors import KNeighborsRegressor
# d)Support vector machine
# from sklearn.svn import SVR
# e) Random forest algorithms
from sklearn.ensemble import RandomForestRegressor
# f) Gradient Boosting
# from sklearn.ensemble import GradientBoostRegressor
# Adaboost
from sklearn.ensemble import AdaBoostRegressor
```

```python
# 4. Cross validation(Hyperparameter Tuning)
# This process
models = []
models.append(('LR', LinearRegression()))
models.append(('DTR', DecisionTreeRegressor()))
models.append(('KNNR', KNeighborsRegressor()))
models.append(('RFR', RandomForestRegressor()))
models.append(('ABR', AdaBoostRegressor()))

# import cross validation and KFOLD
from sklearn.model_selection import cross_val_score,KFold

# create a for loop so that each model is tested in turn
for name,model in models:
  # here we do a 10 spilt K-FOLD
  kfold=KFold(n_splits=10,random_state=42,shuffle=True)
  # We get the results for each fold,Train and Test using the Folds
  cv_result=cross_val_score(model,X_train,Y_train,cv=kfold,scoring='r2')
  # Get the average of all folds for current model in the loop
  print(name,'Results:=',cv_result.mean())

```

```python
# 5. Model Training
# in this process the model learns from the training dataset using the best algorithm
model=LinearRegression()
model.fit(X_train, Y_train)
```

```python
#6. Model evaluation
# this is the process used to determine whether the model is accurate to predict the result
# In regression(numbers) we use the mean absolute error(MAEL)
# To evaluate a model,we will provide the testing features(x_test), then hide testing outcomes(y)
# The model will provide its own predictions(model_predictions),that will be compared against the correct prediction(Y_test)
model_prediction=model.predict(X_test)
print('The Model Predicted: ',model_prediction)

print('\n\n\n')

print('The Correct Answer is: ',Y_test)

print('\n\n')

# Therefore, we will import a scientific matrix called r2_score which is the ratio between the model_prediction and Y_test results
from sklearn.metrics import r2_score
print('The Model Accuracy is: ',r2_score(Y_test,model_prediction)*100,'%')
```

```python
# Predicting results
# The model was trained
features=[[280,25,20]]
sales=model.predict(features)
print('The Predicted Sales is: ',sales)
```
